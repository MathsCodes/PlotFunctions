In order to plot functions properly, we need NumPy and Matplotlib. NumPy is the easy way to work with numbers, and Matplotlib is to show the graphs. 

```python
from matplotlib import pyplot as plt
import numpy as np
```

Now we define the **function** you want to plot. (https://numpy.org/doc/stable/reference/routines.math.html)
Play modifying the function type, instead of power, try to change "np.power(x,2) by the next functions: 

    np.cos(x) -> cosinus
    np.exp(x) -> exponential
    np.sin(x) -> sinus
    np.log10(x) -> base 10 logarithm
    np.tan(x) -> tangent

```python
def f(x): 
    return np.power(x,2)
```

It's time to create the data range, for example: from 0 to 50 with step 1. This is your domain. 
    
    If you put the cos, sin or tan, then change the step to 0.1 or less.
    
You also can play modifying the range. Remember, this "x" contains all de numbers that will go inside the function. It's a list of numbers!! 

```python
x = np.arange(0,100,1)

#Don't pay attention to the next line
print(f'List of numbers: {list(x)}')
```

In order to plot the functions, we need to use the "matplotlib" library. You don't need to modify anything. In order you are curious, you can check other properties in the following link: https://matplotlib.org/stable/gallery/pyplots/pyplot_simple.html#sphx-glr-gallery-pyplots-pyplot-simple-py

```python
plt.plot(x,f(x))
```
